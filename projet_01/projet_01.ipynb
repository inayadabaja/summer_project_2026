# Projet 01 :  SQL - ✈️ interroger une base de réservation de vols 

## But de ce projet 

**Situation :**

En attendant ton vol à l'aéroport de Paris Charles de Gaulle, tu te glisses dans la peau d'un analyste face au système de réservations de l'aéroport...

Dans ce projet, tu vas construire ta propre base de données puis l'interroger avec requêtes simples, des agrégations, jointures et des sous-requêtes... ce qui va finir par attirer l'attention du directeur de l'aéroport... je n'en dis pas plus 😉

Ce premier projet permet de réviser (ou peut-être découvrir ?!) le SQL, le langage universel des bases de données.

Car avant de se lancer dans des projets plus compliqués, il est indispensable de s'échauffer avec les bases (les questions SQL étant fréquentes en entretien d'embauche, il est toujours bon de s'entrainer régulièrement)

Let's go! :)
