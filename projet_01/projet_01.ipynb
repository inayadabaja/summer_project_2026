```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS flights(
    flight_id INTEGER PRIMARY KEY,
    flight_number TEXT NOT NULL UNIQUE,
    departure_airport TEXT NOT NULL,
    arrival_airport TEXT NOT NULL,
    departure_time DATETIME NOT NULL,
    arrival_time DATETIME NOT NULL,
    aircraft_type TEXT NOT NULL
    );
""")

conn.commit()
print("Table 'flights' créée avec succès dans la base de données SQLite en mémoire.")
```

```python
from datetime import datetime, timedelta

airports = [
    ("Paris Charles de Gaulle", "CDG", 0),
    ("London Heathrow", "LHR", 1),
    ("Madrid Barajas", "MAD", 2),
    ("Rome Fiumicino", "FCO", 2),
    ("Amsterdam Schiphol", "AMS", 1),
    ("Frankfurt Airport", "FRA", 1),
    ("Lisbon Airport", "LIS", 2),
    ("Barcelona El Prat", "BCN", 2),
    ("Zurich Airport", "ZRH", 1),
    ("Geneva Airport", "GVA", 1),
    ("New York JFK", "JFK", 8),
    ("Dubai International", "DXB", 7),
    ("Istanbul Airport", "IST", 4),
    ("Casablanca Mohammed V", "CMN", 3),
    ("Tokyo Haneda", "HND", 13),
    ("Montreal Trudeau", "YUL", 8),
    ("Doha Hamad", "DOH", 6),
    ("Cairo International", "CAI", 4),
    ("Athens Airport", "ATH", 3),
    ("Vienna Airport", "VIE", 2)
]

aircraft_types = [
    "Airbus A320",
    "Airbus A321",
    "Airbus A330",
    "Airbus A350",
    "Boeing 737",
    "Boeing 777",
    "Boeing 787",
    "Embraer E190"
]

flights_data = []

start_date = datetime(2026, 8, 1, 6, 0)

for i in range(1, 201):
    departure = airports[0]  # Paris CDG
    arrival = airports[i % len(airports)]

    # Pour éviter CDG -> CDG
    if arrival[1] == "CDG":
        arrival = airports[(i + 1) % len(airports)]

    departure_time = start_date + timedelta(hours=i * 2)
    arrival_time = departure_time + timedelta(hours=arrival[2], minutes=30)

    flight = (
        i,
        f"AF{1000 + i}",
        departure[1],
        arrival[1],
        departure_time.strftime("%Y-%m-%d %H:%M:%S"),
        arrival_time.strftime("%Y-%m-%d %H:%M:%S"),
        aircraft_types[i % len(aircraft_types)]
    )

    flights_data.append(flight)

print(f"{len(flights_data)} vols générés.")
```

```python
cursor.executemany("""
INSERT INTO flights (
    flight_id,
    flight_number,
    departure_airport,
    arrival_airport,
    departure_time,
    arrival_time,
    aircraft_type
)
VALUES (?, ?, ?, ?, ?, ?, ?);
""", flights_data)

conn.commit()

print("200 vols insérés avec succès dans la table 'flights'.")
```
